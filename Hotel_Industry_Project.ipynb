## Data preparation for Hotel Statistic.xlsx

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import data on excel to dataframe.
dfhs = pd.read_excel('Hotel Statistics.xlsx',header = 5) # Used "header" to choose which rows to start from.
dfhs.head(10)
```

```python
# Removes the blank entry.
dfhs = dfhs.drop([0],axis = 0)
dfhs.head(10)
```

```python
# Slice out the necessary data (e.g. 2020 Jan to 2020 Jun).
dfhs = dfhs.iloc[0:6,:]
dfhs
```

```python
# Select the desired columns.
dfhs = dfhs.loc[:,['Time','Hotel Room Revenue(Thousand Dollars)',
                   'Standard Average Hotel Occupancy Rate *(Per Cent)',
                   'Available Room-nights *(Number)']]
dfhs
```

```python
# Renaming of the selected columns.
dfhs.columns = ['Month','Hotel Room Revenue (in $1,000)',
                'Hotel Occupancy Rate (%)','No. of Available Room-nights']
dfhs
```

```python
# Divide "No. of Available Room-nights" column by 1,000 to derive a new column.

dfhs.insert(len(dfhs.columns),'No. of Available Room-nights (in 1,000)',
             (dfhs['No. of Available Room-nights']/1000).to_numpy(),True)
dfhs
```

```python
# Removing "No. of Available Room-nights" column from dataframe.
dfhs = dfhs.drop(['No. of Available Room-nights'],axis = 1)
dfhs
```

```python
# Check for NaN values.
dfhs.isnull().sum()
```

## Data preparation for Int Visitor Length Of Stay.xlsx

```python
dflos = pd.read_excel('Int Visitor Length Of Stay.xlsx',header = 6)
dflos.head(10)
```

```python
# Slice out the necessary data (e.g. 2020 Jan to 2020 Jun).
dflos = dflos.iloc[0:6,:]
dflos
```

```python
# Select the desired columns.
# "Unnamed: 0" refers to Time
# "Unnamed: 1" refers to Total International Visitor Arrivals(Number)
# "Unnamed: 16" refers to Average Length Of Stay (Days)(Day)

dflos = dflos.loc[:,['Unnamed: 0','Unnamed: 1',
                     '15 Days & Over(Number)','Unnamed: 16']]
dflos
```

```python
# Introduce proper names to selected columns.
dflos.columns = ['Month','Total Intl Visitor',
                'No. of Intl Visitor (15 Days & Over)','Avg Length Of Stay (Days)']
dflos
```

```python
# Insert a new variable called "No. of Intl Visitor (Under 15 Days)" to the dataframe.
# "No. of Intl Visitor (Under 15 Days)" = "Total International Visitor" - "No. of Intl Visitor (15 Days & Over)"

dflos.insert(len(dflos.columns),'No. of Intl Visitor (Under 15 Days)',
             (dflos['Total Intl Visitor'] - 
              dflos['No. of Intl Visitor (15 Days & Over)']).to_numpy(),True)
dflos
```

```python
# Divide selected columns by 1,000 to derive new columns.

# No. of Intl Visitor (Under 15 Days)
dflos.insert(len(dflos.columns),'No. of Intl Visitor (in 1,000) <Under 15 Days>',
             (dflos['No. of Intl Visitor (Under 15 Days)']/1000).to_numpy(),True)

# No. of Intl Visitor (15 Days & Over)
dflos.insert(len(dflos.columns),'No. of Intl Visitor (in 1,000) <15 Days & Over>',
             (dflos['No. of Intl Visitor (15 Days & Over)']/1000).to_numpy(),True)

# Total Intl Visitor
dflos.insert(len(dflos.columns),'Total Intl Visitor (in 1,000)',
             (dflos['Total Intl Visitor']/1000).to_numpy(),True)
dflos
```

```python
# Select the desired column.
dflos = dflos.loc[:,['Month','No. of Intl Visitor (in 1,000) <Under 15 Days>',
                     'No. of Intl Visitor (in 1,000) <15 Days & Over>','Avg Length Of Stay (Days)',
                     'Total Intl Visitor (in 1,000)']]
dflos
```

```python
# Check for NaN values.
dflos.isnull().sum()
```

## Data preparation for Outbound Departures Of Singapore Residents.xlsx

```python
dfsing = pd.read_excel('Outbound Departures Of Singapore Residents.xlsx',header = 6)
dfsing.head(10)
```

```python
# Slice out the necessary data (e.g. 2020 Jan to 2020 Jun).
dfsing = dfsing.iloc[0:6,:2]
dfsing
```

```python
# Renaming of the columns.
dfsing.columns = ['Month','Total Departures of Singaporean']
dfsing
```

```python
# Divide "Total Departures of Singaporean" column by 1,000 to derive a new column.

dfsing.insert(len(dfsing.columns),'Total Departures of Singaporean (in 1,000)',
             (dfsing['Total Departures of Singaporean']/1000).to_numpy(),True)
dfsing
```

```python
# Removing "Total Departures of Singaporean" column from dataframe
dfsing = dfsing.drop(['Total Departures of Singaporean'],axis = 1)
dfsing
```

```python
# Check for NaN values.
dfsing.isnull().sum()
```

## Combined Dataset

```python
# Combines all the dataframe into one
dfs = [df.set_index(['Month']) for df in [dfhs, dflos, dfsing,dfcov]]

dfcomb = pd.concat(dfs, axis=1).reset_index()
dfcomb
```

```python
# Reset the index numbers to start from 1.
dfcomb = dfcomb.set_index([pd.Index([i for i in range(1,len(dfcomb)+1)])])
dfcomb
```

## Data Visualization of the Combined Dataset

```python
plt.figure(figsize=(10,10))
plt.rcParams.update({'font.size':19})
plt.style.use('bmh')


months = ('Jan','Feb','Mar','Apr','May','Jun') 

# Plot line graph for "Total Departures of Singaporean (in 1,000)".
plt.plot(months,dfcomb['Total Departures of Singaporean (in 1,000)'],
         marker= 'o',
         markersize= 10, 
         color= '#32a8a0',
         label= 'Departures of Singaporean')

# Plot line graph for "Total of Intl Visitor (in 1,000)".
plt.plot(months,dfcomb['Total Intl Visitor (in 1,000)'],
         marker= 'o',
         markersize= 10,
         color= '#1d4ade',
         label= 'International Visitors')

# Plot line graph for "Total COVID-19 Cases (in 1,000)".
plt.plot(months,dfcomb['Total COVID-19 Cases (in 1,000)'],
         marker= 'X',
         markersize= 12,
         linestyle= '--',
         color= '#d41111',
         label= 'COVID-19 Cases')

plt.xlabel('2020')
plt.ylabel('Total (in 1,000)')
plt.yscale('symlog') # Scaling used for better viewing to showcase comparison.

plt.title('Trend of International Visitors,\nDepartures of Singaporean & COVID-19 Cases',fontsize=22)
plt.legend(loc='best')

# plt.savefig('graph1.png', bbox_inches='tight')

plt.show()
```

```python
# plt.rcdefaults()
plt.figure(figsize=(10,8))
plt.rcParams.update({'font.size':19})

x_axis = (dfcomb['Month'])

# Convert the raw values into percentage.
totals = [i+j for i,j in zip(dfcomb['No. of Intl Visitor (in 1,000) <Under 15 Days>'], 
                             dfcomb['No. of Intl Visitor (in 1,000) <15 Days & Over>'])]

# Create Bars for No. of Visitors under 15 Days
under15days = [i / j * 100 for i,j in zip(dfcomb['No. of Intl Visitor (in 1,000) <Under 15 Days>'], totals)]

# Create Bars for No. of Visitors over 15 Days
over15days = [i / j * 100 for i,j in zip(dfcomb['No. of Intl Visitor (in 1,000) <15 Days & Over>'], totals)]


barWidth = 0.55
months = ('Jan','Feb','Mar','Apr','May','Jun')

# Plotting of bar for No. of Visitors under 15 Days
plt.bar(x_axis, under15days, 
        color='#348abd', 
        edgecolor='black', 
        width=barWidth,
        label= '% of Intl Visitor (Under 15 Days)')

# Plotting of bar for No. of Visitors over 15 Days
plt.bar(x_axis, over15days, 
        bottom=under15days,
        color='#a82a44', 
        edgecolor='black', 
        width=barWidth,
        label= '% of Intl Visitor (15 Days & Over)')

plt.xticks(x_axis, months)
plt.xlabel('2020')
plt.ylabel('International Visitor (%)')

plt.title('Proportion of Total International Visitor \n(Under 15 Days vs 15 Days & Over)',fontsize=25)
plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)

# plt.savefig('graph2.png', bbox_inches='tight')

plt.show()
```
